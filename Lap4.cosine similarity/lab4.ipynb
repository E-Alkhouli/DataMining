```python
import pandas as pd
```

```python
data = pd.read_csv('./cost_revenue_ready.csv') 
```

```python
x = pd.DataFrame(data, columns=['production_budget_usd']) # Features
y = pd.DataFrame(data, columns=['worldwid_gross_usd']) # Target

y
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.3)
plt.title('Production Budget vs Worldwide Gross')
plt.xlabel('Production Budget (USD)')
plt.ylabel('Worldwide Gross (USD)')
plt.ylim(0, 3000000000)  # Set y-axis limit for better visibility
plt.xlim(0, 450000000)  # Set x-axis limit for better
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# LinearRegression can done the job by one object

regression = LinearRegression()
regression.fit(x, y)  # Fit the model
## Slope Coefficients
```

```python
regression.coef_ # Coefficients of the model (slope) (theta 1) 

# it means there is a positive correlation between production budget and worldwide gross
# for each 1 USD increase in production budget, the worldwide gross increases by 3.1 USD on average
# logically, it is true
# good news for movie producers
```

```python
regression.intercept_  # Intercept of the model (theta 0)

# it is -7.2 million USD
# it means if the production budget is 0, the worldwide gross will be -7.2 million USD 
# lose ! # it is not logical, but it is a mathematical model
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.3)

plt.plot(x, regression.predict(x), color='red', linewidth=4)  # Add regression line

plt.title('Production Budget vs Worldwide Gross')
plt.xlabel('Production Budget (USD)')
plt.ylabel('Worldwide Gross (USD)')
plt.ylim(0, 3000000000)  # Set y-axis limit for better visibility
plt.xlim(0, 450000000)  # Set x-axis limit for better
plt.show()
```

```python
regression.score(x, y) # Score of the model (R^2)
```
